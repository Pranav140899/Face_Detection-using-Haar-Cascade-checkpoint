#### Notes
1) Haarcascade files can detect single or multiple occurences of an object from an image of varying shapes and sizes.

2) <b>detectMultiScale()</b> - Detects single or multiple occurrence of an object of different sizes in the input image. The detetced features are returned as coordinates.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
face_cascade = cv2.CascadeClassifier('E:\\Img_datasets\\haarcascade_files\\haarcascade_frontalface_alt.xml')
eye_cascade = cv2.CascadeClassifier('E:\\Img_datasets\\haarcascade_files\\haarcascade_eye.xml')
```

```python
def face_detect(img):
    faces = face_cascade.detectMultiScale(img)
    print(faces)
    for x,y,w,h in faces:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    plt.imshow(img)
    plt.show()
```

```python
a1 = cv2.imread('kneau.jpg')
plt.imshow(a1)
plt.show()
```

```python
def eye_detect(img):
    eyes = eye_cascade.detectMultiScale(img)
    print(eyes)
    for ex,ey,ew,eh in eyes:
        img = cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
    plt.imshow(img)
    plt.show()
```

```python
a1 = cv2.imread('kneau.jpg')
plt.imshow(a1)
plt.show()
```

```python
eye_detect(a1)
```

```python
eye_detect(a2)
```

```python
def detect_face_eye(img):
    faces = face_cascade.detectMultiScale(img)
    for x,y,w,h in faces:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        face_boundary = img[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(face_boundary)
        for ex,ey,ew,eh in eyes:
            cv2.rectangle(face_boundary,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
    plt.imshow(img)
    plt.show()
```

```python
a1 = cv2.imread('kneau.jpg')
plt.imshow(a1)
plt.show()
```

```python
detect_face_eye(a1)
```

```python
a2 = cv2.imread('face2.jpg')
a2 = cv2.cvtColor(a2,cv2.COLOR_BGR2RGB)
plt.imshow(a2)
plt.show()
```

```python
detect_face_eye(a2)
```
